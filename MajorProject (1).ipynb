```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import random
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier

df = pd.read_csv("memory.csv")


# Prints no. of rows and columns of a DataFrame
df.shape
```

```python
# computes various summary statistics, excluding NaN values
df.describe()
```

```python
# This will plot a simple bar chart
plt.bar(df['Algorithm'], df['Memory used (KB)'])

plt.title("Memory Dataset")
plt.legend(["bar"])
plt.show()
```

```python
df.hist()
  
# show plot
plt.show()
```

```python


# Creating plot
fig = plt.figure(figsize =(10, 7))
plt.pie(df['Memory used (KB)'], labels = df['Algorithm'])

# show plot
plt.show()
```
